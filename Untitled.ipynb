```python
import pandas as pd


def normalize_ticker(ticker):
    """
    Convert ticker to standard format.
    """
    if pd.isna(ticker):
        return None

    return str(ticker).strip().upper()


def normalize_year(year):
    """
    Convert year values to integer year.
    """
    if pd.isna(year):
        return None

    year = str(year).strip()

    if year.startswith("FY"):
        return int("20" + year[-2:])

    try:
        return int(float(year))
    except:
        return None
```

```python
print(normalize_ticker("infy"))
print(normalize_ticker(" INFY "))
print(normalize_ticker("adanient"))

print(normalize_year("FY24"))
print(normalize_year("2024"))
print(normalize_year(2024.0))
```

```python
from pathlib import Path
import pandas as pd

RAW_DATA_DIR = Path("data/raw")

HEADER_MAP = {
    "companies.xlsx": 1,
    "stock_prices.xlsx": 0,
    "balancesheet.xlsx": 1,
    "cashflow.xlsx": 1,
    "profitandloss.xlsx": 1,
    "analysis.xlsx": 1,
    "documents.xlsx": 1,
    "prosandcons.xlsx": 1,
    "financial_ratios.xlsx": 1,
    "market_cap.xlsx": 1,
    "peer_groups.xlsx": 1,
    "sectors.xlsx": 1,
}

def load_excel_file(file_name):
    """
    Load Excel file from data/raw using correct header row.
    """
    file_path = RAW_DATA_DIR / file_name

    if not file_path.exists():
        raise FileNotFoundError(f"File not found: {file_path}")

    header = HEADER_MAP.get(file_name, 0)

    df = pd.read_excel(file_path, header=header)

    return df


companies = load_excel_file("companies.xlsx")
stock_prices = load_excel_file("stock_prices.xlsx")

print("companies:", companies.shape)
print("stock_prices:", stock_prices.shape)
```

```python
def validate_primary_key(df, column):
    
    null_count = df[column].isnull().sum()
    
    duplicate_count = df[column].duplicated().sum()
    
    print(f"Column: {column}")
    print(f"Null Values: {null_count}")
    print(f"Duplicate Values: {duplicate_count}")
    
    if null_count == 0 and duplicate_count == 0:
        print("PASS")
    else:
        print("FAIL")
```

```python
def validate_duplicate_rows(df, table_name):
    duplicate_count = df.duplicated().sum()

    print(f"Table: {table_name}")
    print(f"Duplicate Rows: {duplicate_count}")

    if duplicate_count == 0:
        print("PASS")
    else:
        print("FAIL")
```

```python
validate_duplicate_rows(companies, "companies")
```

```python
def validate_missing_values(df, table_name):
    missing = df.isnull().sum()
    missing = missing[missing > 0]

    print(f"Table: {table_name}")

    if len(missing) == 0:
        print("PASS: No missing values")
    else:
        print("FAIL: Missing values found")
        print(missing)
```

```python
validate_missing_values(companies, "companies")
```

```python
def validate_ticker_format(df, column):
    invalid = df[
        df[column].isnull() |
        (df[column].astype(str).str.strip() == "") |
        (~df[column].astype(str).str.match(r"^[A-Z0-9]+$"))
    ]

    print(f"Column: {column}")
    print(f"Invalid Tickers: {len(invalid)}")

    if len(invalid) == 0:
        print("PASS")
    else:
        print("FAIL")
        display(invalid[[column]].head())
```

```python
def validate_numeric_columns(df):

    columns = [
        "face_value",
        "book_value",
        "roe_percentage",
        "roce_percentage"
    ]

    for col in columns:

        valid_rows = df[col].dropna()

        if (valid_rows >= 0).all():
            print(col, "PASS")
        else:
            print(col, "FAIL")
```

```python
def validate_stock_prices(df):
    checks = {
        "open_price > 0": (df["open_price"] > 0).all(),
        "high_price > 0": (df["high_price"] > 0).all(),
        "low_price > 0": (df["low_price"] > 0).all(),
        "close_price > 0": (df["close_price"] > 0).all(),
        "adjusted_close > 0": (df["adjusted_close"] > 0).all(),
        "volume >= 0": (df["volume"] >= 0).all(),
    }

    for rule, result in checks.items():
        print(rule, "PASS" if result else "FAIL")
```

```python
validate_stock_prices(stock_prices)
```

```python
import pandas as pd
import sqlite3
from pathlib import Path
```

```python
companies = load_excel_file("companies.xlsx")
stock_prices = load_excel_file("stock_prices.xlsx")

print(companies.shape)
print(stock_prices.shape)
```

```python
companies["id"] = companies["id"].astype(str).str.strip()
companies["company_name"] = companies["company_name"].astype(str).str.strip()

stock_prices["company_id"] = stock_prices["company_id"].astype(str).str.strip()
stock_prices["date"] = pd.to_datetime(stock_prices["date"]).dt.strftime("%Y-%m-%d")
```

```python
db_path = "data/db/nifty100.db"

conn = sqlite3.connect(db_path)
conn.execute("PRAGMA foreign_keys = ON;")
```

```python
companies.to_sql(
    "companies",
    conn,
    if_exists="append",
    index=False
)

print("Companies loaded")
```

```python
stock_prices.to_sql(
    "stock_prices",
    conn,
    if_exists="append",
    index=False
)

print("Stock prices loaded")
```

```python
companies_count = pd.read_sql("SELECT COUNT(*) AS count FROM companies", conn)
stock_count = pd.read_sql("SELECT COUNT(*) AS count FROM stock_prices", conn)

print(companies_count)
print(stock_count)
```
