```python
import pandas as pd
df = pd.read_csv("../data/processed/master_dataset.csv")

df.head()
```

```python
screeners = PresetScreeners(df)
```

```python
quality = screeners.quality_compounder()

print("Quality Compounder:", len(quality))

quality[
    [
        "company_name",
        "return_on_equity_pct",
        "debt_to_equity",
        "free_cash_flow_cr",
        "operating_profit_margin_pct",
        "composite_quality_score",
    ]
].head(10)
```

```python
value = screeners.value_pick()

print("Value Pick:", len(value))

value[
    [
        "company_name",
        "pe_ratio",
        "pb_ratio",
        "dividend_yield_pct",
        "composite_quality_score",
    ]
].head(10)
```

```python
growth = screeners.growth_accelerator()

print("Growth Accelerator:", len(growth))

growth[
    [
        "company_name",
        "return_on_equity_pct",
        "operating_profit_margin_pct",
        "debt_to_equity",
        "composite_quality_score",
    ]
].head(10)
```

```python
dividend = screeners.dividend_champion()

print("Dividend Champion:", len(dividend))

dividend[
    [
        "company_name",
        "dividend_yield_pct",
        "free_cash_flow_cr",
        "composite_quality_score",
    ]
].head(10)
```

```python
debtfree = screeners.debt_free_bluechip()

print("Debt-Free Blue Chip:", len(debtfree))

debtfree[
    [
        "company_name",
        "debt_to_equity",
        "return_on_equity_pct",
        "composite_quality_score",
    ]
].head(10)
```

```python
turnaround = screeners.turnaround_watch()

print("Turnaround Watch:", len(turnaround))

turnaround[
    [
        "company_name",
        "free_cash_flow_cr",
        "composite_quality_score",
    ]
].head(10)
```

```python
import os

os.makedirs("../output", exist_ok=True)

with pd.ExcelWriter("../output/screener_output.xlsx", engine="openpyxl") as writer:
    quality.to_excel(writer, sheet_name="Quality Compounder", index=False)
    value.to_excel(writer, sheet_name="Value Pick", index=False)
    growth.to_excel(writer, sheet_name="Growth Accelerator", index=False)
    dividend.to_excel(writer, sheet_name="Dividend Champion", index=False)
    debtfree.to_excel(writer, sheet_name="Debt Free Bluechip", index=False)
    turnaround.to_excel(writer, sheet_name="Turnaround Watch", index=False)

print("screener_output.xlsx created successfully.")
```
