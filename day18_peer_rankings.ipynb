```python
import os

os.makedirs("data/processed", exist_ok=True)

peer_df.to_csv(
    "data/processed/peer_percentiles.csv",
    index=False
)

print("peer_percentiles.csv saved.")
```

```python
import sqlite3

conn = sqlite3.connect("data/db/nifty100.db")

peer_df.to_sql(
    "peer_percentiles",
    conn,
    if_exists="replace",
    index=False
)

conn.close()

print("peer_percentiles table created.")
```
