```python
import pandas as pd

# Balance Sheet
balancesheet["year"] = (
    balancesheet["year"]
    .astype(str)
    .str.extract(r'(\d{4})')[0]
)

# Profit & Loss
profitloss["year"] = (
    profitloss["year"]
    .astype(str)
    .str.extract(r'(\d{4})')[0]
)

# Cash Flow (Mar-24 -> 2024)
cashflow["year"] = (
    cashflow["year"]
    .astype(str)
    .str.extract(r'(\d{2})$')[0]
)

cashflow["year"] = "20" + cashflow["year"]

# Convert all to integers
for df in [balancesheet, profitloss, cashflow, ratios, market]:
    df["year"] = pd.to_numeric(df["year"], errors="coerce")
```

```python
print(balancesheet["year"].head())
print(profitloss["year"].head())
print(cashflow["year"].head())
print(ratios["year"].head())
print(market["year"].head())
```

```python
# Remove unnecessary primary key columns

ratios = ratios.drop(columns=["id"], errors="ignore")
market = market.drop(columns=["id"], errors="ignore")
balancesheet = balancesheet.drop(columns=["id"], errors="ignore")
profitloss = profitloss.drop(columns=["id"], errors="ignore")
cashflow = cashflow.drop(columns=["id"], errors="ignore")
```

```python
print(ratios.columns)
print(market.columns)
print(balancesheet.columns)
print(profitloss.columns)
print(cashflow.columns)
```

```python
master = (
    ratios
    .merge(market, on=["company_id", "year"], how="left")
    .merge(profitloss, on=["company_id", "year"], how="left")
    .merge(balancesheet, on=["company_id", "year"], how="left")
    .merge(cashflow, on=["company_id", "year"], how="left")
    .merge(
        companies[["id", "company_name"]],
        left_on="company_id",
        right_on="id",
        how="left"
    )
)

master = master.drop(columns=["id"], errors="ignore")

print(master.shape)
master.head()
```

```python
import os

os.makedirs("../data/processed", exist_ok=True)

master.to_csv("../data/processed/master_dataset.csv", index=False)

print("Saved!")
```

```python
print("Ratios      :", ratios["year"].dtype)
print("Market      :", market["year"].dtype)
print("ProfitLoss  :", profitloss["year"].dtype)
print("BalanceSheet:", balancesheet["year"].dtype)
print("CashFlow    :", cashflow["year"].dtype)
```

```python
# Convert year column to integer in every dataframe

for df in [ratios, market, profitloss, balancesheet, cashflow]:
    df["year"] = (
        pd.to_numeric(df["year"], errors="coerce")
        .fillna(0)
        .astype(int)
    )
```

```python
# Remove unnecessary id columns before merging
ratios = ratios.drop(columns=["id"], errors="ignore")
market = market.drop(columns=["id"], errors="ignore")
profitloss = profitloss.drop(columns=["id"], errors="ignore")
balancesheet = balancesheet.drop(columns=["id"], errors="ignore")
cashflow = cashflow.drop(columns=["id"], errors="ignore")

# Merge datasets
master = (
    ratios
    .merge(market, on=["company_id", "year"], how="left")
    .merge(profitloss, on=["company_id", "year"], how="left")
    .merge(balancesheet, on=["company_id", "year"], how="left")
    .merge(cashflow, on=["company_id", "year"], how="left")
    .merge(
        companies[["id", "company_name"]],
        left_on="company_id",
        right_on="id",
        how="left"
    )
)

# Remove the company id duplicate created by the last merge
master.drop(columns=["id"], inplace=True)

print(master.shape)
master.head()
```

```python
import os

os.makedirs("data/processed", exist_ok=True)

master.to_csv(
    "data/processed/master_dataset.csv",
    index=False
)

print("Saved successfully!")
print(os.path.abspath("data/processed/master_dataset.csv"))
```
