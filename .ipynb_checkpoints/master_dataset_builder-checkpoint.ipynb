```python
import pandas as pd
import numpy as np
```

```python
companies = pd.read_excel(
    "data/raw/companies.xlsx",
    header=1
)

balancesheet = pd.read_excel(
    "data/raw/balancesheet.xlsx",
    header=1
)

profitloss = pd.read_excel(
    "data/raw/profitandloss.xlsx",
    header=1
)

cashflow = pd.read_excel(
    "data/raw/cashflow.xlsx",
    header=1
)

ratios = pd.read_excel(
    "data/raw/financial_ratios.xlsx"
)

market = pd.read_excel(
    "data/raw/market_cap.xlsx"
)
```

```python
print(companies.columns.tolist())

print(balancesheet.columns.tolist())

print(profitloss.columns.tolist())

print(cashflow.columns.tolist())

print(ratios.columns.tolist())

print(market.columns.tolist())
```

```python
profitloss.columns.tolist()
```

```python
cashflow.columns.tolist()
```

```python
# Merge all datasets

master = (
    ratios
    .merge(market, on=["company_id", "year"], how="left")
    .merge(profitloss, on=["company_id", "year"], how="left")
    .merge(balancesheet, on=["company_id", "year"], how="left")
    .merge(cashflow, on=["company_id", "year"], how="left")
    .merge(
        companies[["id", "company_name"]],
        left_on="company_id",
        right_on="id",
        how="left"
    )
)

# Remove duplicate company id column
master.drop(columns=["id"], inplace=True)

print(master.shape)
master.head()
```

```python
print(ratios["year"].head())

print(balancesheet["year"].head())

print(profitloss["year"].head())

print(cashflow["year"].head())

print(market["year"].head())
```

```python
import pandas as pd

# Balance Sheet
balancesheet["year"] = (
    balancesheet["year"]
    .astype(str)
    .str.extract(r'(\d{4})')[0]
)

# Profit & Loss
profitloss["year"] = (
    profitloss["year"]
    .astype(str)
    .str.extract(r'(\d{4})')[0]
)

# Cash Flow (Mar-24 -> 2024)
cashflow["year"] = (
    cashflow["year"]
    .astype(str)
    .str.extract(r'(\d{2})$')[0]
)

cashflow["year"] = "20" + cashflow["year"]

# Convert all to integers
for df in [balancesheet, profitloss, cashflow, ratios, market]:
    df["year"] = pd.to_numeric(df["year"], errors="coerce")
```

```python
print(balancesheet["year"].head())
print(profitloss["year"].head())
print(cashflow["year"].head())
print(ratios["year"].head())
print(market["year"].head())
```

```python
# Remove unnecessary primary key columns

ratios = ratios.drop(columns=["id"], errors="ignore")
market = market.drop(columns=["id"], errors="ignore")
balancesheet = balancesheet.drop(columns=["id"], errors="ignore")
profitloss = profitloss.drop(columns=["id"], errors="ignore")
cashflow = cashflow.drop(columns=["id"], errors="ignore")
```

```python
print(ratios.columns)
print(market.columns)
print(balancesheet.columns)
print(profitloss.columns)
print(cashflow.columns)
```
